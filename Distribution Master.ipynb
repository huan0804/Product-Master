```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import re
```

```python
# Concating raw data
path = r"D:\Python\Python project\Distribution Master"
all_dfs = []
for file in glob.glob(os.path.join(path, "Distribution_Tree_*.xlsx")):
    # Extract year from filename using regex
    filename = os.path.basename(file)
    match = re.search(r"(\d{4})", filename)
    year = match.group(1) if match else None
    # Read the "DATA" sheet
    df_temp = pd.read_excel(file, sheet_name="1.Data dữ liệu")
    # Add year column
    df_temp["File_Year"] = year
    all_dfs.append(df_temp)

# Concatenate all DataFrames
df = pd.concat(all_dfs, ignore_index=True)
```

```python
for columns in df.columns:
    print(f"Null value quantity in {columns}: {df[columns].isnull().sum()}")
```

```python
# Drop unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Rename column
df = df.rename(columns={"Tháng mở": "Open Month","Năm mở": "Open Year","Tháng đóng": "Close Month","Năm đóng": "Close Year"})

# Processing null value in Site store column
df["Site store"] = df["Site store"].fillna("Partnership") 

# Store level: Following channels won't have store level (ONLINE, TGPP, ST, PARTNER)
## Several null value belongs to store level but Channel is CHTT which means they had been closed
df["Store Level"] = df.apply(
    lambda x: "Closed" if pd.isna(x["Store Level"]) and x["Channel"] == "CHTT"
              else ("Other" if pd.isna(x["Store Level"]) and x["Channel"] != "CHTT"
                    else x["Store Level"]), axis=1)

# Transfer Vietnamese value in Trade term into English
df["Trade term"].str.strip()
map_Trade_term = {"Mua đứt bán đoạn":"Buy and sell","Ký gửi":"Consignment"}
df["Trade term"] = df["Trade term"].replace(map_Trade_term)

# Area group null value: if Channel is CHTT then the Area group should be CLOSED, elseif is not CHTT then Area group should be Other
df["Area group"] = df.apply(
    lambda x: "Closed" if pd.isna(x["Area group"]) and x["Channel"] == "CHTT"
              else ("Other" if pd.isna(x["Area group"]) and x["Channel"] != "CHTT"
                    else x["Area group"]), axis=1)
```
